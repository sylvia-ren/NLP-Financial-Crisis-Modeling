```python
import yfinance as yf
import pandas as pd
import numpy as np

# 1. Définir la cible et la période
# KBE est un ETF qui suit l'indice KBW Bank
TICKER = 'KBE'
START_DATE = '2006-12-04'
END_DATE = '2009-01-01'

# 2. Télécharger les données
data = yf.download(TICKER, start=START_DATE, end=END_DATE)

# 3. Calculer les rendements quotidiens (log returns)
data['Log_Returns'] = np.
log(data['Adj Close'] / data['Adj Close'].shift(1))

# 4. Calculer la Volatilité Historique (écart-type des rendements sur 5 jours glissants)
# Annualisation (np.sqrt(252)) car il y a environ 252 jours de trading par an
data['Vol_5J'] = data['Log_Returns'].rolling(window=5).std() * np.sqrt(252)

# Nettoyer (supprimer les premières lignes qui ont des valeurs NaN dues aux calculs)
data = data.dropna()

# 5. Sauvegarder dans le dossier data/
data[['Adj Close', 'Log_Returns', 'Vol_5J']].to_csv('../data/financial_data_KBE.csv')

print("Données financières collectées, traitées et enregistrées dans data/.")
```
